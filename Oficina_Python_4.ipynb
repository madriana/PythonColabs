#**7. Introdução à utilização de biblioteca gráfica (Matplotlib)**

A biblioteca `matplotlib.pyplot` é uma coleção de funções para a **criação de gráficos** 2D a partir de listas. Gráficos comuns podem ser criados com alta qualidade a partir de comandos simples, inspirados nos comandos gráficos do MATLAB.

````
import matplotlib.pyplot as plt
plt.plot( [2,3,1,2,4,5,2] )
plt.title("Meu primeiro gráfico")
plt.show()
````

Se apenas uma lista for passada como parâmetro, o eixo X não está definido. Neste caso ele se inicia em 0 e uma lista é gerada: `[0,1,2,3...]`, do mesmo tamanho da lista (correspondente ao eixo Y) informada.

Se os valores dos dois eixos forem informados, então duas listas serão passadas como parâmetro: eixo X (primeiro argumento) e eixo Y (segundo argumento):


````
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
````

Várias linhas podem ser impressas num mesmo gráfico, e a função `plot` considera pares de listas (eixo x, eixo y). Cada par será uma linha, plotada no gráfico, com uma cor diferente.



````
plt.plot([0.1, 0.2, 0.3, 0.4], [1, 2, 3, 4], [0.1, 0.2, 0.3, 0.4], [1, 4, 9, 16])
plt.show()
````

````
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axis([0, 6, 0, 20])
plt.show()
````

````
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
````

````
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'g^')
plt.axis([0, 6, 0, 20])
plt.show()
````

````

x = [1, 2, 3, 4, 5]
y = [25, 32, 34, 20, 25]
plt.plot(x, y, color='red' , marker='o', markersize=15, linestyle='--', linewidth=3)
````
